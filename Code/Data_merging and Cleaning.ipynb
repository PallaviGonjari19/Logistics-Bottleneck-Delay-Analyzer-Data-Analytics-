```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load your dimension tables
dim_date = pd.read_csv(r"C:\Users\AJINKYA\Desktop\Logistics Bottleneck & Delay Analyzer\Original Datasets\dim_date.csv")
dim_driver = pd.read_csv(r"C:\Users\AJINKYA\Desktop\Logistics Bottleneck & Delay Analyzer\Original Datasets\dim_driver.csv")
dim_hub = pd.read_csv(r"C:\Users\AJINKYA\Desktop\Logistics Bottleneck & Delay Analyzer\Original Datasets\dim_hub.csv")
dim_vehicle = pd.read_csv(r"C:\Users\AJINKYA\Desktop\Logistics Bottleneck & Delay Analyzer\Original Datasets\dim_vehicle.csv")
dim_package = pd.read_csv(r"C:\Users\AJINKYA\Desktop\Logistics Bottleneck & Delay Analyzer\Original Datasets\dim_package.csv")
dim_route = pd.read_csv(r"C:\Users\AJINKYA\Desktop\Logistics Bottleneck & Delay Analyzer\Original Datasets\dim_route.csv")
dim_weather = pd.read_csv(r"C:\Users\AJINKYA\Desktop\Logistics Bottleneck & Delay Analyzer\Original Datasets\dim_weather.csv")
fact_delivery = pd.read_csv(r"C:\Users\AJINKYA\Desktop\Logistics Bottleneck & Delay Analyzer\Original Datasets\fact_delivery.csv")
```

```python
df_merged = fact_delivery.merge(dim_package, on="package_id", how="left") \
                         .merge(dim_driver, on="driver_id", how="left") \
                         .merge(dim_route, on="route_id", how="left") \
                         .merge(dim_vehicle, on="vehicle_id", how="left") \
                         .merge(dim_date, left_on="delivery_date_id", right_on="date_id", how="left") \
                         .merge(dim_weather, on="weather_id", how="left")
```

```python
df_merged.to_csv(r"C:\Users\AJINKYA\Desktop\Logistics Bottleneck & Delay Analyzer\Merged and Cleaned Dataset\delivery_data.csv", index=False)
```

```python
final_data = pd.read_csv(r"C:\Users\AJINKYA\Desktop\Logistics Bottleneck & Delay Analyzer\Merged and Cleaned Dataset\final_logistics_data.csv")
```

```python
final_data
```

```python
# Delivery Status Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data = final_data, x='status', hue='status', palette='Set2', legend=False)
plt.title('Delivery Status Distribution')
plt.xlabel('Status')
plt.ylabel('Number of Deliveries')
plt.tight_layout()
plt.show()
```

```python
# Distribution of Delay Minutes
plt.figure(figsize=(10, 5))
sns.histplot(data=final_data, x='delay_minutes', bins=30, kde=True, color="#FF6F61")
plt.title("Distribution of Delivery Delays (in Minutes)")
plt.xlabel("Delay (Minutes)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Average Delay by Region
plt.figure(figsize=(12, 5))
region_delay = final_data.groupby("region_driver")["delay_minutes"].mean().sort_values(ascending=False)
sns.barplot(x=region_delay.index, y=region_delay.values, palette="coolwarm", hue=None)
plt.title("Average Delay by Region")
plt.xlabel("Region")
plt.ylabel("Avg Delay (Minutes)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Delay by Vehicle Type
plt.figure(figsize=(8, 5))
sns.boxplot(data= final_data, x='vehicle_type', y='delay_minutes', palette="Set3", hue = None)
plt.title("Delay Distribution by Vehicle Type")
plt.xlabel("Vehicle Type")
plt.ylabel("Delay (Minutes)")
plt.tight_layout()
plt.show()
```

```python
# Impact of Weather Condition on Delay
plt.figure(figsize=(12, 5))
sns.boxplot(data=final_data, x='weather_condition', y='delay_minutes', palette="pastel", hue=None)
plt.title("Delay vs Weather Condition")
plt.xlabel("Weather")
plt.ylabel("Delay (Minutes)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
