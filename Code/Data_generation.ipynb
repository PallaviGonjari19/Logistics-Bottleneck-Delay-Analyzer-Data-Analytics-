```python
start_date = datetime(2023, 1, 1)
end_date = datetime(2024, 12, 31)
date_range = pd.date_range(start=start_date, end=end_date)

fog_levels = ['None', 'Low', 'Medium', 'High']
weather_data = []
weather_id = 1

for date in date_range:
    for city in hub_cities:
        temp = round(np.random.normal(loc=30, scale=5), 1)
        precip = max(0, round(np.random.exponential(scale=5), 1))
        fog = random.choices(fog_levels, weights=[0.4, 0.3, 0.2, 0.1])[0]
        if fog != 'None':
            condition = 'Foggy'
        elif precip > 5:
            condition = 'Rainy'
        else:
            condition = random.choice(['Sunny', 'Cloudy'])
        
        weather_data.append([
            weather_id, city, int(date.strftime('%Y%m%d')),
            date.strftime('%H:%M:%S'), temp, precip, fog, condition
        ])
        weather_id += 1

dim_weather = pd.DataFrame(weather_data, columns=[
    'weather_id', 'region', 'date_id', 'time_of_day',
    'temperature_c', 'precipitation_mm', 'fog_intensity', 'weather_condition'
])

dim_weather.to_csv(r"C:\Users\AJINKYA\Desktop\Logistics Bottleneck & Delay Analyzer\Original Datasets\dim_weather.csv", index=False)
```

```python
# Load your dimension tables
dim_date = pd.read_csv(r"C:\Users\AJINKYA\Desktop\Logistics Bottleneck & Delay Analyzer\Original Datasets\dim_date.csv")
dim_driver = pd.read_csv(r"C:\Users\AJINKYA\Desktop\Logistics Bottleneck & Delay Analyzer\Original Datasets\dim_driver.csv")
dim_hub = pd.read_csv(r"C:\Users\AJINKYA\Desktop\Logistics Bottleneck & Delay Analyzer\Original Datasets\dim_hub.csv")
dim_vehicle = pd.read_csv(r"C:\Users\AJINKYA\Desktop\Logistics Bottleneck & Delay Analyzer\Original Datasets\dim_vehicle.csv")
dim_package = pd.read_csv(r"C:\Users\AJINKYA\Desktop\Logistics Bottleneck & Delay Analyzer\Original Datasets\dim_package.csv")
dim_route = pd.read_csv(r"C:\Users\AJINKYA\Desktop\Logistics Bottleneck & Delay Analyzer\Original Datasets\dim_route.csv")
dim_weather = pd.read_csv(r"C:\Users\AJINKYA\Desktop\Logistics Bottleneck & Delay Analyzer\Original Datasets\dim_weather.csv")


# Generate fact_delivery records
fact_data = []
num_records = 50000
np.random.seed(42)

for delivery_id in range(1, num_records + 1):
    package_id = random.choice(dim_package['package_id'])
    driver_id = random.choice(dim_driver['driver_id'])
    vehicle_id = random.choice(dim_vehicle['vehicle_id'])

    route_row = dim_route.sample(1).iloc[0]
    route_id = route_row['route_id']
    origin_location = route_row['origin_location']
    destination_location = route_row['destination_location']
    avg_minutes = route_row['average_time_minutes']

    date_row = dim_date.sample(1).iloc[0]
    delivery_date_id = date_row['date_id']
    delivery_date = pd.to_datetime(date_row['date_actual'])

    # Expected delivery time based on route time and shift
    expected_delivery_time = delivery_date + timedelta(hours=random.randint(8, 20))
    expected_delivery_time += timedelta(minutes=int(avg_minutes))

    # Simulate delay
    delay_choices = [0, 5, 10, 20, 30, 45, 60, 90, 120]
    delay_weights = [0.2, 0.25, 0.2, 0.1, 0.08, 0.05, 0.04, 0.04, 0.04]
    delay_minutes = np.random.choice(delay_choices, p=delay_weights)

    actual_delivery_time = expected_delivery_time + timedelta(minutes=int(delay_minutes))

    if delay_minutes <= 5:
        status = "Delivered"
        on_time_flag = 1
    elif delay_minutes <= 60:
        status = "Delayed"
        on_time_flag = 0
    else:
        status = "Failed"
        on_time_flag = 0

    # Get weather_id by matching region and date
    weather_matches = dim_weather[
        (dim_weather['region'] == origin_location) &
        (dim_weather['date_id'] == delivery_date_id)
    ]
    if not weather_matches.empty:
        weather_id = weather_matches.sample(1).iloc[0]['weather_id']
    else:
        weather_id = np.nan  # fallback

    fact_data.append([
        delivery_id, package_id, driver_id, route_id,
        origin_location, destination_location, vehicle_id,
        status, actual_delivery_time, expected_delivery_time,
        delay_minutes, delivery_date_id, on_time_flag, weather_id
    ])

# Create DataFrame
fact_delivery = pd.DataFrame(fact_data, columns=[
    'delivery_id', 'package_id', 'driver_id', 'route_id',
    'origin_location', 'destination_location', 'vehicle_id',
    'status', 'actual_delivery_time', 'expected_delivery_time',
    'delay_minutes', 'delivery_date_id', 'on_time_flag', 'weather_id'
])

# Save to CSV
fact_delivery.to_csv(r"C:\Users\AJINKYA\Desktop\Logistics Bottleneck & Delay Analyzer\Original Datasets\fact_delivery.csv", index=False)
```
